```python
# Adds dependencies 
import pandas as pd
import numpy as np
```

```python
# Imports data from both csv files 
school_data_to_load = 'Resources/schools_complete.csv'
student_data_to_load = 'Resources/students_complete.csv'
```

```python
# Creates DataFrames out of the csv files
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)
```

```python
# Replaced Thomas High School data for Ninth graders with NaN
student_data_df.loc[(student_data_df['school_name'] == "Thomas High School") & (student_data_df['grade'] == '9th'), "math_score"] = np.nan 
student_data_df.loc[(student_data_df['school_name'] == "Thomas High School") & (student_data_df['grade'] == '9th'), "reading_score"] = np.nan
student_data_df




```

```python
# Creates list of all prefixes and suffixes
student_names = student_data_df['student_name'].tolist()

students_to_fix = []

for name in student_names:
    if len(name.split()) >= 3:
        students_to_fix.append(name)
prefixes = []
for name in students_to_fix:
    if len(name.split()[0]) <= 4:
        prefixes.append(name.split()[0])

suffixes = []
for name in students_to_fix:
    if len(name.split()[-1]) <= 3:
        suffixes.append(name.split()[-1])
```

```python
# List of all professional prefixes and suffixes
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]
```

```python
# Filters out all proffesional prefixes and suffixes from students names and replaces it with nothing
for word in prefixes_suffixes:
    student_data_df['student_name'] = student_data_df['student_name'].str.replace(word,'')
```

```python
# Merges both student and school DataFrames into one DataFrame
school_data_complete_df = pd.merge(student_data_df, school_data_df, on=['school_name', 'school_name'])
```

```python
# Counts the number of students by their STUDENT ID and stores it
student_count = school_data_complete_df["Student ID"].count()
```

```python
# Counts the number of schools and stores it
school_count = school_data_df["school_name"].count()
```

```python
# Uses specifically the school data DataFrame to SUM up all budgets
total_budget = school_data_df["budget"].sum()
```

```python
# Finds the average reading and math scores for all students
average_reading_score = school_data_complete_df['reading_score'].mean()

average_math_score = school_data_complete_df['math_score'].mean()
```

```python
# Finds all students who are passing math or passing reading
passing_math = school_data_complete_df[school_data_complete_df['math_score'] >= 70]

passing_reading = school_data_complete_df[school_data_complete_df['reading_score'] >= 70]
```

```python
# Counts the number of students who are passing math or reading
passing_math_count = passing_math['student_name'].count()

passing_reading_count = passing_reading['student_name'].count()
```

```python
# Percentage of students who are passing math or reading
passing_math_percentage = passing_math_count / float(student_count) * 100

passing_reading_percentage = passing_reading_count / float(student_count) * 100
```

```python
# Shows all the students who are passing math and reading
passing_math_reading = school_data_complete_df[(school_data_complete_df["math_score"] >= 70) & (school_data_complete_df["reading_score"] >= 70)]
```

```python
# Counts how many students are passing both math and reading
overall_passing_math_reading_count = passing_math_reading['student_name'].count()
```

```python
# Percentage of students who are passing both math and reading
overall_passing_percentage = overall_passing_math_reading_count / float(student_count) * 100
```

```python
# Summary of the district including Total students/schools/ budget and Average math/ reading scores and Passing percentages math/ reading/ overall
district_summary_df = pd.DataFrame([{"Total Schools": school_count,
          "Total Students": student_count,
          "Total Budget": total_budget,
          "Average Math Score": average_math_score,
          "Average Reading Score": average_reading_score,
          "% Passing Math": passing_math_percentage,
         "% Passing Reading": passing_reading_percentage,
        "% Overall Passing": overall_passing_percentage}])
```

```python
# Formatting the District Summary
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)

district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)

district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.1f}".format)

district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.1f}".format)

district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:.0f}".format)

district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:.0f}".format)

district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].map("{:.0f}".format)
```

```python
# Creates a new list for the order of the column headers
new_column_order = ["Total Schools", "Total Students", "Total Budget","Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]
```

```python
# Uses new column order to restructer headers in district summary
district_summary_df = district_summary_df[new_column_order]

district_summary_df
```
